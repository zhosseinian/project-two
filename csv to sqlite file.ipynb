```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
larc_file = "ILPQCIPLARC_DATA_2020-01-04_1705.csv"
larc_df = pd.read_csv(larc_file)
larc_df.head()
```

```python
hosp_file = "hospital_data_names.csv"
hosp_df = pd.read_csv(hosp_file)
hosp_df.head()
```

```python
larc_hosp_df = pd.merge(larc_df, hosp_df, on = "hospitalid", how = "left")
larc_hosp_df
```

```python
con = sqlite3.connect("larcs.sqlite")
```

```python
larc_hosp_df.to_sql('table_larcs', con=con)
```

```python
cursor = con.execute("SELECT * FROM table_larcs")
```
